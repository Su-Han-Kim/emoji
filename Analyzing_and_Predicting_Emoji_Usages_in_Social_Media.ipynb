# 3. PROBLEM FORMULATION

## Def 1. Emoji category and position
- emoji category(C: 카테고리의 개수) : $$C_i, i \in [0, C) = [0 \leq i < C]$$ 
- fine grained position(P: 트윗의 최대 길이): $$FP_i, i \in [0, P)$$
- coarse grained position(0: initial, 1: middle, 2: final): $$CP_i, i \in [0, 3)$$

## Def 2. mmGRU network
- G= (X, V, U)
- X = l(tweet) x d(textual)
- V = l(tweet) x e(visual)
- U = l(tweet) x f(user demographics)

## Problem 1: Learning Task
- $$f_1:G_1 = (X, V, U) \rightarrow C \& CP$$
- $$f_2:G_2 = (X, V, U, CP) \rightarrow FP$$

# 5. PROPOSED METHOD


- 이모지의 범주와 위치를 예측하기 위해 multitask multimodality gated recurrent unit (mmGRU) 모델을 제안한다.
- Multimodality function: 텍스트, 이미지 및 사용자 인구 통계와 같은 정보를 통합하는 데 사용
- Multitask: 이모지 category와 해당 위치 간의 강력한 상관 관계를 활용하여 두 작업의 성능을 향상시키는 데 도움이 된다. 
- 제안된 모델을 적용하여 데이터 세트에서 온라인 트윗의 이모지 범주와 위치 예측.
- 이모티콘 예측 방법
  - 트윗에서 텍스트, visual 및 사용자 인구 통계 기능 추출. 
  - 세 가지 유형의 기능이 mmGRU 모델로 전송되어 출력 표현을 학습한다. 
  - 이모티콘 카테고리와 coarse-grained position에 대한 예측 결과를 함께 얻기 위해 softmax classifier 사용. 
  - fine-grained position의 경우, 예측된 결과를 얻기 위해 다른 softmax classifier와 coarse-grained position의 결과를 사용한다. 
  - 잘 훈련된 모델을 기반으로 원시 트윗에서 대상 이모지와 그들의 위치를 얻을 수 있다.

## 5.1 Model structure

- LSTM referance: https://dgkim5360.tistory.com/entry/understanding-long-short-term-memory-lstm-kr?category=912687
- 직사각형 박스는 GRU의 셀을 나타내며, LSTM 셀과  비교하여 리셋 게이트와 업데이트 게이트의 2개의 게이트만 포함합니다.
- reset gate: 새 입력을 이전 메모리와 결합하는 방법을 결정 
- update gate: 이전 메모리의 양을 지정
- GRU의 방정식
- reset vector $$r_t = \sigma(W_{xr}x_t+W_{hr}h_{t-1})$$
- update vector $$z_t = \sigma(W_{xz}x_t+W_{hz}h_{t-1})$$
- candidate output vector $$\tilde{h_t} = tanh(W_{xr}x_t+W_{ho}h_{t-1})$$
- output vector $$h_t = (1-z_t)⊙h_{t-1} + z_t ⊙ \tilde{h_t}$$
- Wxr ,Whr ,Wxz ,Whz: the weight matrix parameters 
- σ is the sigmoid function and tanh is a hyperbolic tangent.
- 텍스트 함수의 경우 단어 임베딩 방법을 사용하여 저차원 벡터로 변환
$$X = x_i , i ∈ [1, t] \rightarrow \tilde{X} = \tilde{x_i}, i \in [1, t]$$

## 5.2. Model learning

As for category and coarse-grained position, we calculate the loss function by incorporating the loss values of both two prediction tasks and train to minimize the value of the loss function J1. As for the fine-grained position, we calculate the loss function J2 alone. The training process uses Adam optimization [7] to perform the classifier-to-neural network feedback and feeds back into the input word vector layer.
